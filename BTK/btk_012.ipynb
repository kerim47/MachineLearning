```python
import pandas as pd
satislar = pd.read_csv('http://bilkav.com/satislar.csv')
```

```python
df = satislar.copy()
df.head()
```

```python
X = df[['Aylar']]
Y = df[['Satislar']]
X
Y
```

```python
from sklearn.model_selection import train_test_split
bolme     = 0.33
rnd_state = 0
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=bolme, random_state=rnd_state)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
```

```python
import pickle
dosya = 'model.kayit'
pickle.dump(lin_reg, open(dosya, 'wb'))
kayit_lin_reg = pickle.load(open(dosya, 'rb'))
```
